### <font color='blue'>Criação do Tópico Kafka</font>

```python
# Imports
from confluent_kafka.admin import AdminClient, NewTopic
from confluent_kafka.error import KafkaException
```

```python
def dsa_cria_topico(topic_name, num_partitions, replication_factor):

    # Cliente o cliente com os endereços dos brokers
    admin_client = AdminClient({'bootstrap.servers': 'kafka1:29092,kafka2:29093,kafka3:29094'})

    # Define um novo tópico
    novo_topico = NewTopic(topic_name,
                           num_partitions = num_partitions,
                           replication_factor = replication_factor)
    
    # Cria o tópico no Kafka
    fs = admin_client.create_topics([novo_topico])

    # Loop para verificar se o tópico foi criado
    for topic, f in fs.items():
        try:
            f.result()
            print(f"Tópico {topic} criado com {num_partitions} partições e fator de replicação de {replication_factor}")
        except KafkaException as e:
            print(f"Falha ao criar o tópico {topic}: {e}")
```

```python
dsa_cria_topico('dsa_p8_topico', 6, 2)
```
